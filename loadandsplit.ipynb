```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter

pdf_data = PyPDFLoader("NOTEBOOK LM ASSIGNMENT.pdf").load()

full_data = "\n".join([doc.page_content for doc in pdf_data])

text_split = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=100, chunk_overlap=0)
char_split = text_split.split_text(full_data)
char_split
```

```python
from langchain_community.document_loaders import PyPDFLoader
split_text = PyPDFLoader("NOTEBOOK LM ASSIGNMENT.pdf")
textex = split_text.load()
full_text = "\n".join([doc.page_content for doc in textex])
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=100, chunk_overlap=0)

char_split = text_splitter.split_text(full_text)
char_split
```
